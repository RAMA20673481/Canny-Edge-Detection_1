```python
low=int(max(0,0.7*medium))
low
```

```python
high=int(max(255,1/3*medium))
high
```

```python
img = cv2.imread('SEC.jpg')
if img is None:
    print("Error: Image not found!")
else:
    blur_img = cv2.blur(img, ksize=(3,3))
    if len(blur_img.shape) == 2: 
        plt.imshow(blur_img, cmap='gray')
        plt.title("Blurred Grayscale Image")
    else: 
        plt.imshow(cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB))
        plt.title("Blurred Color Image")
    plt.axis('off')
    plt.show()
```

```python
canny_edge_blur=cv2.Canny(blur_img,threshold1=low,threshold2=high)
plt.imshow(canny_edge_blur,cmap='gray')
```
